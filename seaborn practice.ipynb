```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```
