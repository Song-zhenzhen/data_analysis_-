```python
import scipy.io
import numpy as np
import pandas as pd
```

```python
#只取其中一个导联作为展示数据，一共15个视频片段
data2 = np.ones(32000)
for i in range(1,16):
    data = scipy.io.loadmat('jingjing_20140629.mat')['jj_eeg%d'%i]
    data1 = data[0][0:32000]
    data2 = np.vstack((data2,data1))
data2.shape
```

```python
data = data2[1:]
data.shape
```

```python
import matplotlib.pyplot as plt
import matplotlib
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
data = min_max_scaler.fit_transform(data)
data[0]
   
```

```python
plt.xlim(0,32000)
plt.ylim(-0.5,1.5)
plt.xlabel('eeg',fontsize='x-large')
plt.ylabel('movie',fontsize='x-large')
x=np.arange(0,32000)
plt.plot(x,data[0],marker='.',ls='-',ms=0.5,mfc='w',lw=0.3,label='wave',c='g')
plt.show()
```
